```python
df_clean = df[['Sample name', 'DropPlate Position', 'A260 Concentration (ng/ul)',
       'A260/A280', 'A260/A230', 'Raw A340 (10mm)']]
df_clean
```

### Outliers for 'A260/A280' 

Acceptance criteria = 1.8 to 2.2


```python
import plotly.graph_objects as go
import plotly.offline as po
import plotly.express as px

fig = px.scatter(
                df_clean,
                x=df_clean['DropPlate Position'],
                y=df_clean["A260/A280"],
                hover_name = df_clean['Sample name'],
                title="A260/280 Vs Drop plate position(Acceptance criteria 1.8 to 2.2)",
    
                )

#Set colors for contamination
fig.update_traces(mode='markers',
                  
    marker=dict(
        size=8,
       
        color=(
            (df_clean['A260/A280'] > 2.2) |
            (df_clean['A260/A280'] < 1.8)
        ).astype('float'),
        colorscale=[[0, 'green'], [1, 'red']]
    )
)

fig.add_hline(y=1.8)
fig.add_hline(y=2.2)

fig.update_traces(name='A260/A280 not contaminated', showlegend = True)
fig.show()
```

### Outliers for 'A260/A230'
Acceptance criteria = 2.0 to 2.3

```python
import plotly.graph_objects as go
import plotly.offline as po
import plotly.express as px

fig = px.scatter(
                df_clean,
                x=df_clean['DropPlate Position'],
                y=df_clean["A260/A230"],
                hover_name = df_clean['Sample name'],
                title="A260/230 Vs Drop plate position(Acceptance criteria 2.0 to 2.3)",
    
                )

#Set colors for contamination
fig.update_traces(mode='markers',
                  
    marker=dict(
        size=8,
       
        color=(
            (df_clean['A260/A230'] > 2.3) |
            (df_clean['A260/A230'] < 2.0)
        ).astype('float'),
        colorscale=[[0, 'green'], [1, 'red']]
    )
)

fig.add_hline(y=2.0)
fig.add_hline(y=2.3)

fig.update_traces(name='A260/A230 not contaminated', showlegend = True)
fig.show()
```
