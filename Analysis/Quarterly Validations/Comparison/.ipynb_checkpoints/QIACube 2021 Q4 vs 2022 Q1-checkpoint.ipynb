```python
df_combined = df_q4_21.append(df_q1_22)
df_combined = df_combined.append(df1_q3_21)
df_combined
```

```python
from colorama import Fore, Back, Style

mean_group = df_combined.groupby(['Quarter','Machine']).mean()
std_group = df_combined.groupby(['Quarter','Machine']).std()
print(Back.YELLOW   + "Mean values for each machines in each Quarter")
mean_group
```

```python
from colorama import Fore, Back, Style

print(Back.YELLOW   + "SD values for each machines in each Quarter")
std_group
```

```python
import plotly.express as px

df_conc_mean = df_combined.groupby(['Quarter','Machine']).agg(mean_conc=("A260 Concentration (ng/ul)", 'mean'))
df_conc_mean= df_conc_mean.reset_index()

fig = px.bar(df_conc_mean, x="Machine",
           y="mean_conc",
       title = "Mean A260 concentration comparison: QIA 1 Vs QIA 2 (2021 Q4 Vs 2022 Q1)",
            facet_col='Quarter',
            text_auto=True,
                  labels=dict(Machine =" ",mean_conc="Mean : A260 concentration"),
    
)

fig.update_layout(
    title={
        
        'y':0.97,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    font_family="Arial",
    
    title_font_color="crimson",
)


fig.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_yaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))

fig.show()

#fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
```

```python
import plotly.express as px


df_conc_std = df_combined.groupby(['Quarter','Machine']).agg(std_conc=("A260 Concentration (ng/ul)", 'std'))
df_conc_std= df_conc_std.reset_index()

fig = px.bar(df_conc_std, x="Machine",
           y="std_conc",
       title = "Standard Deviation A260 concentration comparison: QIA 1 Vs QIA 2 (2021 Q4 Vs 2022 Q1)",
            facet_col='Quarter',
            text_auto=True,
                  labels=dict(Machine =" ",std_conc="Standard Deviation : A260 concentration"),
    
)

fig.update_layout(
    title={
        
        'y':0.97,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    font_family="Arial",
    
    title_font_color="crimson",
)

fig.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_yaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))

fig.show()

#fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
```

```python
df_yield_mean = df_combined.groupby(['Quarter','Machine']).agg(Mean_Yield=("Yield(ug)", 'mean'))
df_yield_mean= df_yield_mean.reset_index()


fig = px.bar(df_yield_mean, x="Machine",
           y="Mean_Yield",
       title = "Mean Yield comparison: QIA 1 Vs QIA 2 (2021 Q4 Vs 2022 Q1)",
            facet_col='Quarter',
            text_auto=True,
                  labels=dict(Machine =" ",Mean_Yield="Mean of Yield"),
    
)


fig.update_layout(
    title={
        
        'y':0.97,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    font_family="Arial",
    
    title_font_color="crimson",
)


fig.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_yaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))

fig.show()

#fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
```

```python
df_yield_std = df_combined.groupby(['Quarter','Machine']).agg(sd_Yield=("Yield(ug)", 'std'))
df_yield_std= df_yield_std.reset_index()

fig = px.bar(df_yield_std, x="Machine",
           y="sd_Yield",
           facet_col='Quarter',
                  text_auto=True,
                  title="Yield SD comparison: QIA 1 Vs QIA 2 (2021 Q4 Vs 2022 Q1)",
                  labels=dict(Machine =" ",sd_Yield="Standard Deviation of Yield"),
    
)

fig.update_layout(
    title={
        
        'y':0.94,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    font_family="Arial",
    
    title_font_color="crimson",
)

fig.update_xaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_yaxes(tickfont=dict(family='Rockwell', color='crimson', size=14))

fig.show()
```
