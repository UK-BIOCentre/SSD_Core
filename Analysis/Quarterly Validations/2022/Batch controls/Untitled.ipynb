# Mean Value for Quarters

```python
cl_q4_21 = round(df4_21['A260 Concentration (ng/ul)'].mean(),2)
cl_q1_22 = round(df1_22['A260 Concentration (ng/ul)'].mean(),2)

cl_q2_22 = round(df2_22['A260 Concentration (ng/ul)'].mean(),2)
cl_q4_21,cl_q1_22,cl_q2_22
```

## SD , 2SD and 3SD for all Quarters

```python
sd_q4_21=round(df4_21['A260 Concentration (ng/ul)'].std(),2)

sd_q1_22 = round(df1_22['A260 Concentration (ng/ul)'].std(),2)

sd_q2_22 = round(df2_22['A260 Concentration (ng/ul)'].std(),2)
sd_q4_21,sd_q1_22,sd_q2_22
```

```python
# 1 SD
q4_21_sd_pos = cl_q4_21 + sd_q4_21
q4_21_sd_neg = cl_q4_21 - sd_q4_21

q1_22_sd_pos = round(cl_q1_22 + sd_q1_22,2)
q1_22_sd_neg = round(cl_q1_22 - sd_q1_22,2)

q2_22_sd_pos = cl_q2_22 + sd_q2_22
q2_22_sd_neg = cl_q2_22 - sd_q2_22
```

```python
q4_21_2sd_pos = round(cl_q4_21 + 2*sd_q4_21,2)
q4_21_2sd_neg = round(cl_q4_21 - 2*sd_q4_21,2)

q1_22_2sd_pos = round(cl_q1_22 + 2*sd_q1_22,2)
q1_22_2sd_neg = round(cl_q1_22 - 2*sd_q1_22,2)

q2_22_2sd_pos = round(cl_q2_22 + 2*sd_q2_22,2)
q2_22_2sd_neg = round(cl_q2_22 - 2*sd_q2_22,2)

q4_21_3sd_pos = round(cl_q4_21 + 3*sd_q4_21,2)
q4_21_3sd_neg = round(cl_q4_21 - 3*sd_q4_21,2)


q1_22_3sd_pos = round(cl_q1_22 + 3*sd_q2_22,2)
q1_22_3sd_neg = round(cl_q1_22 - 3*sd_q2_22,2)

q2_22_3sd_pos = round(cl_q2_22 + 3*sd_q2_22,2)
q2_22_3sd_neg = round(cl_q2_22 - 3*sd_q2_22,2)
```

```python
data = {'Quarter':['Q4 21','Q1 22','Q2 22'],
       'Mean':[cl_q4_21,cl_q1_22,cl_q2_22],
       '+1 SD':[q4_21_sd_pos,q1_22_sd_pos,q2_22_sd_pos],
       '-1 SD':[q4_21_sd_neg,q1_22_sd_neg,q2_22_sd_neg],
       '+2 SD':[q4_21_2sd_pos,q1_22_2sd_pos,q2_22_2sd_pos],
       '-2 SD':[q4_21_2sd_neg,q1_22_2sd_neg,q2_22_2sd_neg],
       '+3 SD':[q4_21_3sd_pos,q1_22_3sd_pos,q2_22_3sd_pos],
       '-3 SD':[q4_21_3sd_neg,q1_22_3sd_neg,q2_22_3sd_neg]}
data
```

```python
df_stat = pd.DataFrame(data)
```

```python
df_stat
```

```python

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat.Mean,fillcolor='green',
                    mode='lines+markers',
                    name='Mean'))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['+1 SD'],
                    mode='lines+markers',
                    name='+1 SD'))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['+2 SD'],
                    mode='lines+markers',
                    name='+2 SD'))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['+3 SD'],
                    mode='lines+markers',
                    name='+3 SD'))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['-1 SD'],
                    mode='lines+markers',
                    name='-1 SD'
                    ))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['-2 SD'],
                    mode='lines+markers',
                    name='-2 SD'))
fig.add_trace(go.Scatter(x=df_stat.Quarter, y=df_stat['-3 SD'],
                    mode='lines+markers',
                    name='-3 SD'))
fig.show()
```

# Q4 2021

```python
import plotly.graph_objects as go

print(df_stat.iloc[:1])
fig = go.Figure(data=go.Scatter(y=df['A260 Concentration (ng/ul)'],
           mode='lines+markers'))

fig.add_hline(y=cl_q4_21, line_color = 'green', line_dash="dash",
              annotation_text="Mean Value", 
              annotation_position="top right")

fig.add_hline(y=q4_21_3sd_pos, line_color = 'red',
              annotation_text="LCL (+ 3SD)", 
              annotation_position="top right")
fig.add_hline(y=q4_21_2sd_pos, line_color = 'indianred',line_dash='dot',
              annotation_text="UCL (+2SD)", 
              annotation_position="top right")

fig.add_hline(y=q4_21_sd_pos, line_color = 'magenta',line_dash="dash",
              annotation_text="+ 1 SD", 
              annotation_position="top right")

fig.add_hline(y=q4_21_sd_neg, line_color = 'magenta',line_dash="dash",
              annotation_text="-1 SD", 
              annotation_position="top right")


fig.add_hline(y=q4_21_2sd_neg, line_color = 'indianred',line_dash='dot',
              annotation_text="LCL (- 2SD)", 
              annotation_position="top right")

fig.add_hline(y=q4_21_3sd_neg, line_color = 'red',
              annotation_text="LCL (- 3SD)", 
              annotation_position="top right")
fig.update_layout(title= 'Q4 2024 Negative BC Analysis',
                  yaxis_title="A260 Concentration")

fig.show()

fig.write_html("Q4_21.html")
```
