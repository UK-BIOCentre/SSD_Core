```python

import os
import glob
import pandas as pd

#import numpy as np
```

```python
PATH = 'C:\\Users\\hema.ramasamy\\OneDrive - Uk Biocentre\\SSD_Core\\Analysis\\Quarterly Validations\\2022\\Q1 Chemagen\\Data'
#PATH = os.getcwd()
temps_csvs = glob.glob(os.path.join(PATH, "*.csv"))     

df_for_each_csv = (pd.read_csv(f) for f in temps_csvs)
df = pd.concat(df_for_each_csv, ignore_index = True)
```

```python
df['Yield(ug)'] = (df['Concentration'] * df['Volume'])/1000 # Total sample volume
```

```python
df['Date'] = pd.to_datetime(df['ProcessedDateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
df['Date'] = df['Date'].dt.date
```

```python
df_chem_q1_q4 = df[['Date','Project','Method', 'OD260/280',\
                    'OD260/230','Concentration','Yield(ug)']]
df_chem_q1_q4.to_csv(r'C:\Users\hema.ramasamy\OneDrive - Uk Biocentre\SSD_Core\Analysis\Quarterly Validations\Comparison\Combined_Chemagen_Q1_22_Q4_21.csv')
```
