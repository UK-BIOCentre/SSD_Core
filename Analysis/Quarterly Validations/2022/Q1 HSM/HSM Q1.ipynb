```python
conc_mean = round(df_q1_22_chem_C['A260 Concentration (ng/ul)'].mean(),2)
conc_std = round(df_q1_22_chem_C['A260 Concentration (ng/ul)'].std(),2)

mean_260_280 = round(df_q1_22_chem_C['A260/A280'].mean(),2)
std_260_280 = round(df_q1_22_chem_C['A260/A280'].std(),2)

mean_260_230 = round(df_q1_22_chem_C['A260/A230'].mean(),2)
std_260_230 = round(df_q1_22_chem_C['A260/A230'].std(),2)

raw_340_mean = round(df_q1_22_chem_C['Raw A340 (10mm)'].mean(),2)
raw_340_std = round(df_q1_22_chem_C['Raw A340 (10mm)'].std(),2)

print(f'A260 concentration Mean:  {conc_mean} and standard deviation : {conc_std}')
print(f'A260/A280 Mean :  {mean_260_280} and standard deviation: {std_260_280}')
print(f'A260/A230 Mean : {mean_260_230} and standard deviation :  {std_260_230}')
print(f'Raw A340 Mean : {raw_340_mean} and standard deviation : {raw_340_std}')
```

```python
X = ['A260 concentration', 'A260/A280','A260/A230','Raw A340']
Mean = [conc_mean,mean_260_280,mean_260_230,raw_340_mean]
std = [conc_std, std_260_280,std_260_230,raw_340_std]
  
X_axis = np.arange(len(X))

plt.barh(X_axis - 0.2, Mean, 0.4, label = 'Mean')

#plt.barh(X_axis + 0.2, std, 0.4, label = 'Standard Dev')
  
plt.yticks(X_axis, X)
plt.ylabel("Measures")
#plt.ylabel("Number of Students")
plt.title("Mean of Measures")
plt.legend()

for index, value in enumerate(Mean):
    plt.text(value, index,
             str(value))
plt.show()
```
