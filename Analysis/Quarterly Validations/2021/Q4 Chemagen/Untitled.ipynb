```python
import pandas as pd,numpy as np
```

```python
df_cns = pd.read_csv(r'NIHR-NBC-RTB-CNS-DNA-Extraction-Report-202205191330.csv')
df_rdc = pd.read_csv(r'NIHR-NBC-RDC-DNA-Extraction-Report-202205191312.csv')
df_cvd = pd.read_csv(r'NIHR-NBC-RTB-CVD-DNA-Extraction-Report-202205191314.csv')
df_gen = pd.read_csv(r'NIHR-NBC-RTB-GEN-DNA-Extraction-Report-202205191312.csv')
df_ibd = pd.read_csv(r'NIHR-NBC-RTB-IBD-DNA-Extraction-Report-202205191314.csv')
df_imi = pd.read_csv(r'NIHR-NBC-RTB-IMI-DNA-Extraction-Report-202205191251.csv')
df_naf = pd.read_csv(r'NIHR-NBC-RTB-NAF-DNA-Extraction-Report-202205191311.csv')
```

```python
dfs = [df_cns,df_rdc,df_cvd,df_gen,df_ibd,df_imi,df_naf]

df = pd.concat(dfs)
df.head()
```

```python
df['Yield(ug)'] = (df['Concentration'] * df['Volume'])/1000 # Total sample volume 
df.head(2)
```

```python
df['Date'] = pd.to_datetime(df['ProcessedDateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
df['Date'] = df['Date'].dt.date
```

```python
df_chem_q1_22 = df[['Date','Project','Method', 'OD260/280', 'OD260/230','Concentration','Yield(ug)']]
df_chem_q1_22.to_csv('Combined_Chemagen_Q1_2022.csv')
```
