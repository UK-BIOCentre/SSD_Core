```python
import pandas as pd
import numpy as np

import datetime
from datetime import date,datetime

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hema.ramasamy\OneDrive - Uk Biocentre\SSD_Core\Analysis\Apr 2022\RTB-GEN\NIHR-NBC-RTB-GEN-DNA-Extraction-Report-202205090915.csv')
df
```

```python
df['Date'] = pd.to_datetime(df['ProcessedDateTime'], format='%d/%m/%Y %H:%M:%S')
```

```python
df['Date'] = df['Date'].dt.date
```

```python
df1 = df[['ParticipantID','Date', 'OD260/280', 'OD260/230','Concentration']]
df1
```

```python
df1.isna().sum()
```

```python
df1['OD260/280'] = pd.to_numeric(df1['OD260/280'], errors='coerce')

df1 = df1.dropna(subset=['OD260/280'])
df1
```

```python
df1['OD260/230'] = pd.to_numeric(df1['OD260/230'], errors='coerce')

df1 = df1.dropna(subset=['OD260/230'])
df1
```

```python
df2 = df1.groupby('Date').count()
df2 = df2.sort_values(by = 'Date').reset_index()
df2
```

```python
df2.sort_values(by = 'Date')
```

```python
ax = sns.barplot(x='Date',
                 y='ParticipantID',
                 data = df2,
                                 )
plt.xticks(rotation=90)
plt.title("Count of OD260/280 VS Date ")
for p in ax.patches:
    ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.5, p.get_height()+0.02))

plt.show()
```

```python
import plotly.express as px

fig = px.scatter(df1, x="Date", y="OD260/280",
                 hover_data=['ParticipantID'],title = 'Contamination : OD260/280',
                
    
        color=(
            (df1['OD260/280'] > 2.04) |
            (df1['OD260/280'] < 1.75)
        ).astype('float'),
        color_continuous_scale=["green", "red"]
        
    )
fig.add_hline(y=1.75, line_width=2, line_color="blue")
fig.add_hline(y=2.04, line_width=2, line_color="blue")


fig.show()
```

```python
cont_df_280 = df1.loc[(df1['OD260/280']< 1.75) | (df1['OD260/280']>2.04)]
cont_df_280.set_index('Date').sort_values('Date')

def highlight_col(x):
    r = 'background-color: yellow'
    df_280 = pd.DataFrame(index=x.index, columns=x.columns)
    df_280.iloc[:, 2] = r
    return df_280   
cont_df_280.to_csv('280.csv')
cont_df_280.style.apply(highlight_col, axis=None)
```

```python
fig = px.scatter(df1, x="Date", y="OD260/230",
                 hover_data=['ParticipantID'],title = 'Contamination OD260/230',
                
    
        color=(
            
            (df1['OD260/230'] <1.5)
        ).astype('int'),
        color_continuous_scale=["green", "red"]
        
    )
fig.add_hline(y=1.5, line_width=2, line_color="blue")
#fig.add_hline(y=2.0, line_width=2, line_color="blue")


fig.show()
```

```python
cont_df_230 = df1.loc[(df1['OD260/230']<1.5)]
cont_df_230.set_index('Date').sort_values('Date')

def highlight_col(x):
    r = 'background-color: yellow'
    df_test = pd.DataFrame(index=x.index, columns=x.columns)
    df_test.iloc[:, 3] = r
    return df_test    
cont_df_230.style.apply(highlight_col, axis=None)
```

```python
df1['Concentration'] = pd.to_numeric(df1['Concentration'], errors='coerce')

df1 = df1.dropna(subset=['Concentration'])
df1
```

```python
fig = px.scatter(df1, x="Date", y="Concentration",
                 hover_data=['ParticipantID'],title = 'Contamination based on Concentration',
                
    
        color=(
            (df1['Concentration'] < 30) #|
           # (df1['OD260/230'] < 2.0)
        ).astype('int'),
        color_continuous_scale=["green", "red"]
        
    )
#fig.add_hline(y=2.3, line_width=2, line_color="blue")
fig.add_hline(y=30, line_width=2, line_color="blue")


fig.show()
```

```python
conc_df= df1.loc[(df1['Concentration']<30)]
conc_df.set_index('Date').sort_values('Date')

def highlight_col(x):
    r = 'background-color: yellow'
    df_c = pd.DataFrame(index=x.index, columns=x.columns)
    df_c.iloc[:, 4] = r
    return df_c    
conc_df.style.apply(highlight_col, axis=None)
```
