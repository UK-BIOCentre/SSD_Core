```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'L:\Sample acceptance\Senior Lab tech info\Controls for CB Lab\Chemagens\Negative Control\Neg control data batch test 20211025_v1.xlsm',sheet_name = 'Chem NC Data')
df
```

```python
df = df[['Date', 'Project', 'Chemagen', 'SOURCE Rack ID', 'Source Tube ID',
       'Output Tube ID', 'Position', 'A260 Concentration (ng/ul)', 'A260/A280',
       'A260/A230', 'Raw A340 (10mm)']]
df
```

```python
cl = round(df['A260 Concentration (ng/ul)'].mean(),2)
cl
```

```python
sd=round(df['A260 Concentration (ng/ul)'].std(),2)
```

```python
ucl = cl + 2*sd
lcl = cl - 2*sd

sd_pos = cl + sd
sd_neg = cl - sd

sd3pos = cl + 3*sd
sd3neg = cl - 3*sd
```

```python
lst =[]

for a in df['A260 Concentration (ng/ul)']:
    lst.append(a)
```

```python
fig, ax = plt.subplots(figsize = (10, 6))
ax.plot(lst, linestyle='-', marker='o', color='blue')



plt.axhline(cl,linestyle='-', marker='o', color='green')
ax.axhline(ucl, color='red')
ax.axhline(lcl, color='red')
ax.axhline(sd_pos, color='aqua')
ax.axhline(sd_neg, color='aqua')
ax.axhline(sd3pos, color='crimson',linestyle='--')
ax.axhline(sd3neg, color='crimson',linestyle='--')

ax.set(xlabel='Sample', ylabel='A260 Concentration')

# Determine the x-axis limits in the chart to attach reference values
left, right = ax.get_xlim()
ax.text(right + 0.5, ucl, "+2SD = " + str("{:.2f}".format(ucl)), color='red')
ax.text(right + 0.5, sd_pos, "+ 1 SD  = " + str("{:.2f}".format(sd_pos)), color='red')
ax.text(right + 0.5, sd_neg, "- 1 SD  = " + str("{:.2f}".format(sd_neg)), color='red')
ax.text(right + 0.5, sd3pos, "+ 3 SD  = " + str("{:.2f}".format(sd_pos)), color='red')
ax.text(right + 0.5, sd3neg, "- 3 SD  = " + str("{:.2f}".format(sd_neg)), color='red')
ax.text(right + 0.5, cl, 'Mean' + " = " + str("{:.2f}".format(cl)), color='green')
ax.text(right + 0.5, lcl, "-2SD = " + str("{:.2f}".format(lcl)), color='red')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(y=df['A260 Concentration (ng/ul)'],
           mode='lines+markers'))

fig.add_hline(y=cl, line_color = 'green', line_dash="dot",
              annotation_text="Mean Value", 
              annotation_position="top right")

fig.add_hline(y=sd3pos, line_color = 'red',
              annotation_text="+3 SD", 
              annotation_position="top right")

fig.add_hline(y=ucl, line_color = 'indianred',line_dash='dot',
              annotation_text="UCL (+2SD)", 
              annotation_position="top right")

fig.add_hline(y=sd_pos, line_color = 'magenta',line_dash="dash",
              annotation_text="+ 1 SD", 
              annotation_position="top right")

fig.add_hline(y=sd_neg, line_color = 'magenta',line_dash="dash",
              annotation_text="-1 SD", 
              annotation_position="top right")


fig.add_hline(y=lcl, line_color = 'indianred',line_dash='dot',
              annotation_text="LCL (- 2SD)", 
              annotation_position="top right")

fig.add_hline(y=sd3neg, line_color = 'red',
              annotation_text="-3 SD", 
              annotation_position="top right")




fig.show()

fig.write_html("Negative_Control_Chemagen_21_22.html")
```
